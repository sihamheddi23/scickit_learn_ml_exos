```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Audio
```

```python
def analyse_audio(path):
  audio,rate = librosa.load(path)
  Audio(data=audio, rate=rate)
  print("number of samples ",audio.size)
  print("frequency ", rate)
  plt.figure(figsize=(10, 4))
  librosa.display.waveshow(audio, sr=rate)
  plt.title('Waveform')
  plt.xlabel('Time (s)')
  plt.ylabel('Amplitude')
  plt.show()
  plt.figure(figsize=(12, 4))
  ft = np.abs(librosa.stft(audio))
  plt.plot(ft);
  plt.title('Spectrum');
  plt.xlabel('Frequency Bin');
  plt.ylabel('Amplitude');
  D = librosa.stft(audio)
  S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
  fig, ax = plt.subplots(figsize=(10, 5))
  img = librosa.display.specshow(S_db,
                                x_axis='time',
                                y_axis='log',
                                ax = ax,
                               )
  fig.colorbar(img, ax=ax, format=f'%0.2f')
```

```python
analyse_audio("/content/hello.mp3")
```

```python
analyse_audio("/content/car.wav")
```
